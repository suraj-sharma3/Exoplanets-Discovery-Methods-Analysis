#### To check the standard deviation in a column

## Standard Deviation Explanation

### Standard deviation is a way to measure how "spread out" or "different" the numbers in a group are from the average (mean). Think of it as a number that tells you how much the values in a set vary from the center.

- If the standard deviation is small, it means most numbers are close to the average.
- If the standard deviation is large, it means the numbers are more spread out, with some being much higher or lower than the average.

#### Example : 

- Imagine you're measuring how long it takes your friends to get to school:

    - Friend 1: 15 minutes
    - Friend 2: 16 minutes
    - Friend 3: 15 minutes
    - Friend 4: 15 minutes
    - Friend 5: 14 minutes

- Here, the times are all very close to the average, so the standard deviation is small.

- Now, imagine these times:

    - Friend 1: 10 minutes
    - Friend 2: 30 minutes
    - Friend 3: 15 minutes
    - Friend 4: 25 minutes
    - Friend 5: 5 minutes

- In this case, the times vary a lot more from the average, so the standard deviation is large.

#### Visual Analogy : 

- Think of the numbers in your data as dots on a piece of paper:

    - Small standard deviation: The dots are tightly clustered around the center.
    - Large standard deviation: The dots are scattered far and wide.

#### Standard Deviation Formula :

- Find the mean of the numbers.
- Subtract the mean from each number to get the differences.
- Square those differences.
- Find the average of those squared differences.
- Take the square root of the result.

```python
data['Num Planets'].std()
```

#### To fill some value in cells of a column where there is NaN value
- Use the fillna() method to fill missing (NaN) values in a specific column of a pandas DataFrame. 
    - df["column_name"] = df["column_name"].fillna(value) # This only modifies the specified column without affecting others.
    - You can replace 0 with any value or logic, such as the column's mean, median, or mode.
    - df["Column1"] = df["Column1"].fillna(df["Column1"].mean()) # Fill NaN with the column's mean
    - You can also provide a string value to the fillna method

- If you call .fillna() on the entire DataFrame, it will attempt to fill all the NaN values across all columns : 
    - filled_df = df.fillna(0) # Fill NaN values in the entire DataFrame with a specific value (e.g., 0)
    - df.fillna(0, inplace = True) # Fill NaN values in the entire DataFrame with a specific value (e.g., 0), the changes will be made in the made in the original dataframe df as inplace is True

```python
data['Orbital Period Days'] = data['Orbital Period Days'].fillna(0.0)
```

```python
nan_count_orb_per_days = data['Orbital Period Days'].isnull().sum()
print(f"Number of NaN values in Orbital Period Days column : {nan_count_orb_per_days}")
```
